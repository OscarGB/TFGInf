```python
df = pd.read_parquet("time_dataset.parquet")
```

```python
cor = df.corr().abs()
f = plt.figure(figsize=(19, 15))
plt.matshow(cor, fignum=f.number, cmap='YlOrBr')
plt.xticks(range(cor.shape[1]), cor.columns, fontsize=10, rotation=90)
plt.yticks(range(cor.shape[1]), cor.columns, fontsize=10)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=10)
plt.title('Correlation Matrix', fontsize=16);
```
