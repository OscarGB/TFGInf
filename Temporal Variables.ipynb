# Temporal Variables
Este notebook realiza una modificacion de las variables para incluir datos del pasado y no solo del año actual.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_parquet("dataset.parquet")
```

```python
df = data.copy()
indexNames = df[ df.Year < 1992 ].index
df.drop(indexNames , inplace=True)
df.head()
```

```python
df.columns
```

```python
df["Last Year Municipal"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Municipal"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Municipal)
df.head()
```

```python
df["Last Year Household"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Household"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Household)
df.head()
```

```python
df["Last Year Recovered"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Recovered"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Recovered)
df.head()
```

```python
df["Last Year Recicled"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Recicled"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Recicled)
df.head()
```

```python
df["Last Year Compost"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Compost"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Compost)
df.head()
```

```python
df["Last Year Built"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Last Year Built"] = \
            float(data[(data.COU == c) & (data.Year == y-1)].Built)
df.head()
```

```python
df["Difference Built"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Difference Built"] = \
            float(data[(data.COU == c) & (data.Year == y)].Built) - float(data[(data.COU == c) & (data.Year == y-1)].Built)
df.head()
```

```python
df["Mean Built"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Mean Built"] = \
            np.mean(data[(data.COU == c) & (data.Year < y)].Built)
df.head()
```

```python
df['Last Year Below Secundary'] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), 'Last Year Below Secundary'] = \
            float(data[(data.COU == c) & (data.Year == y-1)]['Below Secundary'])
df.head()
```

```python
df["Difference Below Secundary"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Difference Below Secundary"] = \
            float(data[(data.COU == c) & (data.Year == y)]['Below Secundary']) - float(data[(data.COU == c) & (data.Year == y-1)]['Below Secundary'])
df.head()
```

```python
df['Last Year Population'] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), 'Last Year Population'] = \
            float(data[(data.COU == c) & (data.Year == y-1)]['Population'])
df.head()
```

```python
df["Difference Population"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Difference Population"] = \
            float(data[(data.COU == c) & (data.Year == y)]['Population']) - float(data[(data.COU == c) & (data.Year == y-1)]['Population'])
df.head()
```

```python
df['Last Year Tourism'] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), 'Last Year Tourism'] = \
            float(data[(data.COU == c) & (data.Year == y-1)]['Tourism'])
df.head()
```

```python
df["Difference Tourism"] = 0
for c in df.COU.unique():
    for y in df.Year.unique():
        df.loc[(df.COU == c) & (df.Year == y), "Difference Tourism"] = \
            float(data[(data.COU == c) & (data.Year == y)]['Tourism']) - float(data[(data.COU == c) & (data.Year == y-1)]['Tourism'])
df.head()
```
